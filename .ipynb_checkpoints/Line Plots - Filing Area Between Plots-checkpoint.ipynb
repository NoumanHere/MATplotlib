# Filing area on Line Plots

```python
import pandas as pd
from matplotlib import pyplot as plt

#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AL
```

```python
data = pd.read_csv('data.csv')
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
pydev_salaries = data['Python']
jsdev_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color = '#444444',linestyle = '--', label = "All Developers")
plt.plot(ages, pydev_salaries,linestyle = 'solid', label = "Python Developers")

overall_median_salary = 57287

plt.fill_between(ages,pydev_salaries,overall_median_salary,
                 where = (pydev_salaries > overall_median_salary),
                 interpolate=True,alpha = 0.25, color = 'blue')
plt.fill_between(ages,pydev_salaries,overall_median_salary,
                 where = (pydev_salaries < overall_median_salary),
                 interpolate=True,alpha = 0.25, color = 'orange')

plt.legend()
plt.title("Median Salary (USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```

```python
plt.plot(ages, dev_salaries, color = '#444444',linestyle = '--', label = "All Developers")
plt.plot(ages, pydev_salaries,linestyle = 'solid', label = "Python Developers")

overall_median_salary = 57287

plt.fill_between(ages,pydev_salaries,dev_salaries,
                 where = (pydev_salaries > dev_salaries),
                 interpolate=True,alpha = 0.25, color = 'blue', label = 'Above AVG')
plt.fill_between(ages,pydev_salaries,dev_salaries,
                 where = (pydev_salaries <= dev_salaries),
                 interpolate=True,alpha = 0.25, color = 'orange', label = 'Below AVG')

plt.legend()
plt.title("Median Salary (USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```
