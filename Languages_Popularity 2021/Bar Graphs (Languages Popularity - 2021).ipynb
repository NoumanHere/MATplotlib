# Bar Charts and Loading Data from csv 

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import csv
from collections import Counter
```

```python
# With pandas
data = pd.read_csv('data.csv')
# ids = data['Responder_id']
lang_responses = data['LanguageHaveWorkedWith']

language_counter = Counter()
for response in lang_responses:
    language_counter.update(response.split(';'))
    
    
languages = []
popularity = []
common_langs = language_counter.most_common(15)
for item in common_langs:
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

```

```python
plt.xkcd()
# print(plt.style.available)
# plt.style.use('')
width = 0.5
plt.barh(languages,popularity,width, color = 'black')
plt.title('Most Popular Languages (2021)')
plt.xlabel('Popularity')
plt.tight_layout()
plt.show()
```
